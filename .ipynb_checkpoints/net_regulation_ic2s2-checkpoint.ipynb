```python
def get_similar_strings(df, col):
    dict_similar = {}
    for word in tqdm(df[col]):
        words = sources[df[col] != word][col]
        similar_words = difflib.get_close_matches(word, words, len(words), 0.6)
        dict_similar[word] = similar_words
    return dict_similar
```

```python
#LDA model
texts = []
for i in tqdm.tqdm(range(len(data['lem_text']))):
    text = data['lem_text'][i]
    texts.append(nltk.word_tokenize(str(text)))
```

```python
# Create Dictionary
id2word = corpora.Dictionary(texts)
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
# Set up log to external log file
logging.basicConfig(filename='lda_model_ic2s2.log', format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# Build LDA model
lda_model_emo = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=8, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=1,
                                           alpha=0.001,
                                           per_word_topics=True,
                                           minimum_probability=0)
```

```python
for i in range(8):
    print(lda_model_emo.show_topic(i, 10))
```

```python
def get_topics_for_docs(corpus, k, clean_texts):
    '''get most probable topics for corpus documents'''
    docs_train_ids = []
    topics_train = []
    for i in tqdm.tqdm(list(range(len(corpus)))):
        seen_doc = corpus[i]
        vector = lda_model_emo[seen_doc]
        probs = []
        for t in list(range(k)):
            probs.append(vector[0][t][1])
        docs_train_ids.append(i)
        topics_train.append(probs.index(max(probs)))
    df_topics = pd.DataFrame({'clean_text': clean_texts, 'topic': topics_train})
    return df_topics
```

```python
grammar = ('''
    AS: {<A=m><S>}
    SS: {<S><S>}
    S1: {<S>}
    ''')

chunkParser = nltk.RegexpParser(grammar)
```

```python
dfs = []
i = 0
for text in tqdm.tqdm(lem_text):
    i = i + 1
    num_text = "text {}".format(i)
    tagged = nltk.pos_tag(nltk.word_tokenize(text), lang='rus')
    tree = chunkParser.parse(tagged)
    as_tag = []
    ss_tag = []
    s_tag = []
    for subtree in tree.subtrees():
        if subtree.label() == 'AS': as_tag.append(" ".join(untag(subtree)).lower())
        if subtree.label() == 'SS': ss_tag.append(" ".join(untag(subtree)).lower())
        if subtree.label() == 'S1': s_tag.append(" ".join(untag(subtree)).lower())
    as_tag_df = pd.DataFrame({'term': as_tag})
    as_tag_df['tag'] = 'as_tag'
    
    ss_tag_df = pd.DataFrame({'term': ss_tag})
    ss_tag_df['tag'] = 'ss_tag'
    
    s_tag_df = pd.DataFrame({'term': s_tag})
    s_tag_df['tag'] = 's_tag'

    df = as_tag_df.append(ss_tag_df)
    df = df.append(s_tag_df)
    df['num_text'] = num_text
    
    dfs.append(df)

final_df = pd.concat(dfs)
```

```python
import re

def search(target, text, context=11):
    # It's easier to use re.findall to split the string, 
    # as we get rid of the punctuation
    words = re.findall(r'\w+', text)

    matches = (i for (i,w) in enumerate(words) if w.lower() == target)
    for index in matches:
        if index < context //2:
            yield words[0:context+1]
        elif index > len(words) - context//2 - 1:
            yield words[-(context+1):]
        else:
            yield words[index - context//2:index + context//2 + 1]
```

```python
nearest = []
for text in tidy_text:
    nearest.append(list(search('блокировка', str(text))))
```

```python
n = 1
reg_df = []
for scope in nearest:
    if len(scope) == 0:
        reg = pd.DataFrame({'word': ['none']})
        reg['id'] = n
        reg_df.append(reg)
        n = n + 1
    else:
        for i in scope:
            reg = pd.DataFrame({'word': i})
            reg['id'] = n
            reg_df.append(reg)
            n = n + 1           
    
reg_df = pd.concat(reg_df)
```
